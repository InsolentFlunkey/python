**Collatz Conjecture** - Start with a number *n > 1*. Find the number of steps it takes to reach one using the following process: If *n* is even, divide it by 2. If *n* is odd, multiply it by 3 and add 1.

```python
while True:
    try:
        num = int(input("Enter an integer greater than zero: "))
    except:
        print("Input value must be an integer")
    else:
        if num <= 0:
            print("Input must be positive")
        else:
            break

ct = 0

while num != 1:
    if num % 2 == 0:
        # num is even, so divide by 2
        num = num / 2
    else:
        # num is odd, so multiply by 3 and add 1
        num = num * 3 + 1
    
    print("Num = {}".format(int(num)))
    ct += 1
    
print("Reached 1 in {} turns".format(ct))
    
```
