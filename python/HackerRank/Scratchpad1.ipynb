```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the viralAdvertising function below.
def viralAdvertising(n):
    shared = 5
    total = 0
    for ct in range(1,n+1):
        liked = int(shared / 2)
        shared = liked * 3
        total += liked
        
        print("Day: {}, shares: {}, Liked: {}, Total:{}".format(ct,shared,liked,total))
    return total

if __name__ == '__main__':

    n = int(input())

    result = viralAdvertising(n)
    print(result)

```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the saveThePrisoner function below.
def saveThePrisoner(people, candy, start):
    remaining = candy
    if candy > people:
        remaining = candy % people
    
    if start > 1:
        start -= 1
    else:
        start = people

    last = start + remaining
    if last > people:
        last -= people
    return last

if __name__ == '__main__':

    t = int(input())

    for t_itr in range(t):
        nms = input().split()

        n = int(nms[0])

        m = int(nms[1])

        s = int(nms[2])

        result = saveThePrisoner(n, m, s)
        
        print(result)


# Test case
# 352926151 380324688 94730870
# answer should be 122129406
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the circularArrayRotation function below.
def circularArrayRotation(a, k, queries):
    shifts = k % len(a)
    retArr = []
    
    if shifts > 0:
        for ct in range(shifts):
            a.insert(0,a.pop())
    
    for query in queries:
        retArr.append(a[query])
        
    return retArr

if __name__ == '__main__':

    nkq = input().split()

    n = int(nkq[0])

    k = int(nkq[1])

    q = int(nkq[2])

    a = list(map(int, input().rstrip().split()))

    queries = list(map(int, input().rstrip().split()))

    #for _ in range(q):
     #   queries_item = int(input())
      #  queries.append(queries_item)

    result = circularArrayRotation(a, k, queries)

    print("\nOutput:")
    print('\n'.join(map(str, result)))
    
# Sample input
# 3 2 3
# 1 2 3
# 0
# 1
# 2

# Expected Output
# 2
# 3
# 1
```
